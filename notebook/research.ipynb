```python
# user who rated atleast 200 books
most_rated = ratings_df['id'].value_counts() >= 200
```

### Mergering most_rated with books_df

```python
ratings_with_books = most_rated.merge(books_df)
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.isnull().sum()
```

```python
ratings_with_books.shape
```

```python
ratings_with_books.describe()
```

```python
most_rated_books = ratings_with_books.groupby('title')['ratings'].count().reset_index()
```

```python
most_rated_books.rename(columns={'ratings':'num_of_ratings'}, inplace=True)
```

```python
most_rated_books.head()
```

### Merge `ratings_with_books` with `most_rated_books`

```python
final_ratings = ratings_with_books.merge(most_rated_books)
```

```python
final_ratings.head()
```

```python
final_ratings = final_ratings[final_ratings['num_of_ratings'] >= 50]
```

```python
final_ratings.head()
```

```python
final_ratings.drop_duplicates(['id', 'title'], inplace=True)
```

```python
final_ratings.head()
```

```python
book_pivot = final_ratings.pivot_table(index='title', columns='id', values='ratings')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

### Converting pivot table into a CSR-Matrix

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
book_pivot.iloc[237,:]
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1, -1), n_neighbors=6)
```

### Recommender function

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1, -1), n_neighbors=6)
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"You searched `{book_name}`\n")
                print(f"Suggestions:\n")
            else:
                print()
```
